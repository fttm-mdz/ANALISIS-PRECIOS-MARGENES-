```python
#lipieza de nombres de columnas
def limpiar_columnas(cols):
    nuevas_cols = []
    for col in cols:
        col = col.lower()
        col = unicodedata.normalize("NFKD", col).encode("ascii", "ignore").decode("utf-8")
        col = re.sub(r"[()]", "", col)
        col = re.sub(r"\s+", "_", col)
        col = re.sub(r"_+", "_", col)
        col = col.strip("_")
        nuevas_cols.append(col)
    return nuevas_cols

ventas_df.columns = limpiar_columnas(ventas_df.columns)
ventas_df.columns

ventas_df
```
